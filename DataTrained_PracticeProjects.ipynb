```python
#  Practice Project 2
#  Mushroom Project

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder


df=pd.read_csv('C:\ProgramData\mushrooms.csv')
df.head()

df.shape

df.isnull().sum()

%matplotlib inline

sns.countplot(df['class'])

le=LabelEncoder()

df=df.apply(le.fit_transform)

x=df.drop(['class'],axis=1)
y=df['class']

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=10)

dt=DecisionTreeClassifier(max_depth=5)
dt.fit(x_train,y_train)

y_predicted_dt=dt.predict(x_test)

from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_predicted_dt)

#from sklearn.metrics import classification_report
#classification_report(y_test,y_predicted_dt)



```

```python
lr=LogisticRegression()
lr.fit(x_train,y_train)
y_predicted_lr=lr.predict(x_test)

accuracy_score(y_test,y_predicted_lr)
```

```python
rfl=RandomForestClassifier(max_depth=5)
rfl.fit(x_train,y_train)
y_predicted_rfl=lr.predict(x_test)

accuracy_score(y_test,y_predicted_rfl)
```

```python
NOTE:Out of LogisticRegression,DecisionTreeClassifier and RandomForestClassifier DecisionTreeClassifier has highest accuracy_score.Hence
    it is the best out of all three.
```

```python
# Practice Project 1 
# World Happiness Report


import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv('C:\ProgramData\happiness_score_dataset.csv')
df.head(5)
```

```python
plt.figure(figsize=(14,6))
sns.scatterplot(y=df['Region'],x=df['Freedom'],hue=df['Region'])
plt.legend(loc='upper left')
plt.title("Region Vs FREEDOM")
```

```python
Clearly above plot represents that 'Freedom' for Western Europe is the 'most' and 'least' for 'sub-saharan africa'
```

```python
plt.figure(figsize=(14,6))
sns.scatterplot(y=df['Region'],x=df['Happiness Score'],hue=df['Region'])
plt.legend(loc='upper left')
plt.title('Region Vs HAPPINESS SCORE')
```

```python
Clearly above plot depicts that Happiness Score for Western Europe is the maximum and minimum in sub-saharan africa
```

```python
plt.figure(figsize=(14,6))
sns.scatterplot(y=df['Region'],x=df['Health (Life Expectancy)'],hue=df['Region'])
plt.legend(loc='upper left')
plt.title('REGION Vs Health (Life Expectancy)')
```

```python
Clearly above plot depicts that Health(Life expectancy) for Western Europe is the maximum and minimum in sub-saharan africa
```

```python
plt.figure(figsize=(14,6))
plt.title('REGION')
plt.xlabel('Region')
plt.ylabel('Health (Life Expectancy)')
plt.xticks(rotation=30,ha='right')
plt.bar(df['Region'],df['Health (Life Expectancy)'])
```

```python
plt.figure(figsize=(14,6))
df2=df.head(10)
plt.title('Country Vs Health (Life Expectancy)')
plt.xlabel('Country')
plt.ylabel('Health (Life Expectancy)')
plt.xticks(rotation=30,ha='right')
plt.bar(df2['Country'],df2['Health (Life Expectancy)'])
```
